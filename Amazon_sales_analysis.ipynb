# Data cleaning and data preparation

```python
#Seperating numerical and categorical features
```

```python
numerical_features=[feature for feature in df.columns if df[feature].dtype!='O']
```

```python
categorical_features=[feature for feature in df.columns if df[feature].dtype=='O']
```

```python
categorical_features
```

```python
#changing type of ship-postal-code and Date to appropriate format
df['ship-postal-code'].astype('float64')
df['Date']=pd.to_datetime(df['Date'])
```

```python
#Extracting month from Date and adding it at 1st index
```

```python
df.insert(1,'Month',df['Date'].dt.month_name())
```

```python
df
```

```python
df.isnull().sum()
```

```python
df[numerical_features].isnull().sum()
```

```python
df[categorical_features].isnull().sum()
```

```python
# Checking unique values for fulfilled-by,ship-country,currency
```

```python
df['fulfilled-by'].unique()
```

```python
df['ship-country'].unique()
```

```python
df['currency'].unique()
```

```python
#Dropping columns fulfilled-by,ship-country,currency,Unnamed: 22
#fulfilled-by has only one category i.e Easy Ship with no other relationship
#currency contains INR currency i.e Indian Currency
#shipping country is India
#Unnamed: 22 contains more NaN values and it contains data that cannot be determined or identified
```

```python
df.info()
```

```python
df.drop(['fulfilled-by','Unnamed: 22','currency','ship-country','Sales Channel '],inplace=True,axis=1)
```

```python
df.isnull().sum()
```

```python
#Checking duplicate values
df[df.duplicated(['Order ID','ASIN'],keep=False)]
```

```python
#Dropping duplicate values
df.drop_duplicates(['Order ID','ASIN'],inplace=True,ignore_index=True)
```

```python
df.head()
```

```python
# Checking NaN values
df.isnull().sum()
```

```python
# Filling NaN values
df['Courier Status'].fillna('Missing',inplace=True)
df['Amount'].fillna(0,inplace=True)
df['ship-city'].fillna('Missing',inplace=True)
df['ship-state'].fillna(0,inplace=True)
df['ship-postal-code'].fillna('Missing',inplace=True)
df['promotion-ids'].fillna(0,inplace=True)
df['ship-postal-code'].fillna(0.0,inplace=True)
```

```python
df.isnull().sum()
```

```python
# Renaming columns Order ID to order_id, Courier Status to courier_status,Qty to Quantity,promotion-ids to promotion
df.rename(columns={'Order ID':'order_id','Courier Status':'courier_status','Qty':'Quantity','promotion-ids':'promotion'},inplace=True)
```

```python
df.info()
```
